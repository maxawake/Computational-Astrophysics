```python
import numpy
from matplotlib.pyplot import *

%matplotlib inline





mev = 1.6e-15
NA = 6.022e+23



#density
rho = 160.


#rates (MS)
R0 = 8.1e-20*rho/2.
R1 = 0.011*rho
R2 = 2.2e-10*rho/2.
R3 = 3.6e-16*rho
R4 = 1.2e-3
R5 = 1.2e-15*rho
R6 = 6.9e-19*rho
R7 = 5.7e-3
R8 = 3.0e-14*rho




#useful vectors
J = numpy.zeros((10,10))
C = numpy.zeros((10))
dY = numpy.zeros((10))

yr = 3.14e+7

################################################################

def Cfunc(Yn1,Yn,dt,C):


    #p
    C[0] = (Yn1[0]-Yn[0])/dt    -      (-2*Yn1[0]**2*R0 -Yn1[0]*Yn1[1]*R1 + 2*Yn1[2]**2*R2 -Yn1[0]*Yn1[3]*R3 -Yn1[0]*Yn1[5]*R5 -Yn1[0]*Yn1[6]*R6 -Yn1[0]*Yn1[8]*R8)

    #d
    C[1] = (Yn1[1]-Yn[1])/dt    -      (Yn1[0]**2*R0 -Yn1[0]*Yn1[1]*R1)

    #he3
    C[2] = (Yn1[2]-Yn[2])/dt    -      (-2*Yn1[2]**2*R2 + Yn1[0]*Yn1[1]*R1)

    #c12
    C[3] = (Yn1[3]-Yn[3])/dt    -      (-Yn1[0]*Yn1[3]*R3 + Yn1[0]*Yn1[8]*R8)

    #n13
    C[4] = (Yn1[4]-Yn[4])/dt    -      (-Yn1[4]*R4 + Yn1[0]*Yn1[3]*R3)

    #c13
    C[5] = (Yn1[5]-Yn[5])/dt    -      (-Yn1[0]*Yn1[5]*R5 + Yn1[4]*R4)

    #n14
    C[6] = (Yn1[6]-Yn[6])/dt    -      (-Yn1[0]*Yn1[6]*R6 + Yn1[0]*Yn1[5]*R5)

    #o15
    C[7] = (Yn1[7]-Yn[7])/dt    -      (-Yn1[7]*R7 + Yn1[0]*Yn1[6]*R6)

    #n15
    C[8] = (Yn1[8]-Yn[8])/dt    -      (-Yn1[0]*Yn1[8]*R8 +Yn1[7]*R7)

    #he4
    C[9] = (Yn1[9]-Yn[9])/dt    -      (Yn1[2]**2*R3 + Yn1[0]*Yn1[8]*R8)	


    return C



def Jfunc(Yn1,dt,J):




    #p
    J[0,0] = 1./dt -   (-4.*Yn1[0]*R0 -Yn1[1]*R1 -Yn1[3]*R3 -Yn1[5]*R5 -Yn1[6]*R6 -Yn1[8]*R8)
    J[0,1] = - (-Yn1[0]*R1)
    J[0,2] = - (4*Yn1[2]*R2)
    J[0,3] = -(-Yn1[0]*R3)
    J[0,5] = -(-Yn1[0]*R5)
    J[0,6] = -(-Yn1[0]*R6)
    J[0,8] = -(-Yn1[0]*R8)

    #d
    J[1,0] = - (2*Yn1[0]*R0-Yn1[1]*R1)
    J[1,1] = 1./dt - (-Yn1[0]*R1)
    
    #he3
    J[2,0] = - (Yn1[1]*R1)
    J[2,1] = - (Yn1[0]*R1)
    J[2,2] = 1./dt - (-4.*Yn1[2]*R2) 

    #c12
    J[3,0] = -  (-Yn1[3]*R3 + Yn1[8]*R8 )
    J[3,3] = 1./dt - (-Yn1[0]*R3)
    J[3,8] = - (Yn1[0]*R8)

    #n13
    J[4,0] = -  (Yn1[3]*R3)
    J[4,3] = - (Yn1[0]*R3)
    J[4,4] = 1./dt-(-R4)


    #c13
    J[5,0] = -  (-Yn1[5]*R5)
    J[5,4] = - (R4)
    J[5,5] = 1./dt - (-Yn1[0]*R5)

    #n14
    J[6,0] = - (-Yn1[6]*R6 + Yn1[5]*R5 )
    J[6,5] = - (Yn1[0]*R5)
    J[6,6] = 1./dt - (-Yn1[0]*R6)

    #o15
    J[7,0] = -  (Yn1[6]*R6)
    J[7,6] = - (Yn1[0]*R6)
    J[7,7] = 1./dt - (- R7)

    #n15
    J[8,0] = -  (-Yn1[8]*R8)
    J[8,7] = - (R7)
    J[8,8] = 1./dt - (-Yn1[0]*R8)


    #he4
    J[9,0] = - (Yn1[8]*R8)
    J[9,2] = - (2*Yn1[2]*R2)
    J[9,8] = - (Yn1[0]*R8)
    J[9,9] = 1./dt


    return J



def int_1step(Yn1,Yn,dt,C,J):

    C_i = Cfunc(Yn1,Yn,dt,C)
    J_i = Jfunc(Yn1,dt,J)
    J_i_inv = numpy.linalg.inv(J_i) 
    dY = - numpy.matmul(J_i_inv, C_i)
    Yn1 = Yn1 + dY	
    dY_max = numpy.abs(numpy.max(dY/Yn1))


    return Yn1,Yn,dt,C,J,dY_max


################################################################

X  = numpy.array((7.095E-01,2.756E-05,8.464E-05,2.337E-03,0.,2.845E-05,8.130E-04,0.,3.197E-06,2.750E-01))
labels = ['p','d',r'$^3He$',r'$^{12}C$',r'$^{13}N$',r'$^{13}C$',r'$^{14}N$',r'$^{15}O$',r'$^{15}N$',r'$^{4}He$']



A  = numpy.array((1.,2.,3.,12.,13.,13.,14.,15.,15.,4.))
Yn = X/A
Yn1 = numpy.zeros((10))

#eps
eps = 1.e-3
it_max = 5
tmax = 3.14e+17

#initial time
t=0.
dt = 0.1
Y = []
time = []
fout = 1

z = 0
#main loop
while(t<tmax):


    if z%fout==0:

            Y.append(A*Yn)
            time.append(t)
        
    if z//fout > 1000:

            break


    Yn1[:] = Yn[:]

    


    dY_max = 2*eps

    it = 0.
    while dY_max > eps:


            Yn1,Yn,dt,C,J,dY_max = int_1step(Yn1,Yn,dt,C,J)
            it +=1.

            if it>it_max:
                print(it)
                dt = dt/10.
                it = 0.
                Yn1[:] = Yn[:]
                dY_max = 2*eps




    rel_err = numpy.max(abs((Yn1-Yn)/Yn1))
    dt = 0.1*(1./rel_err)*dt


    t = t + dt
    Yn = Yn1

    z = z+1

    
    
time1 = numpy.array(time)  
Y1 = numpy.array(Y)










```
