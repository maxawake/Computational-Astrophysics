# Exercise 2: Finite Differences

## a)

```python
def forward(f, x, h):
    return (f(x + h) - f(x)) / h

def backward(f, x, h):
    return (f(x) - f(x - h)) / h

def central(f, x, h):
    return (f(x + h) - f(x - h)) / (2 * h)
```

## b)

```python
dxs = [10**(-i) for i in range(0,16)]
print("Step sizes:", dxs)

forwards = [np.abs(1 - forward(np.exp, 0, dx)) for dx in dxs]
backwards = [np.abs(1 - backward(np.exp, 0, dx)) for dx in dxs]
centrals = [np.abs(1 - central(np.exp, 0, dx)) for dx in dxs]

plt.plot(dxs, forwards, label="Forward")
plt.plot(dxs, backwards, "--", label="Backward")
plt.plot(dxs, centrals, "-.", label="Central")

plt.title("Finite difference quotients for $\\frac{d}{dx}\exp(0)=1$")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Step size $\Delta x$")
plt.ylabel("Relative error $|1-\Delta x/\Delta y|$")
plt.legend()
```

# c)

One can observe here, that the relative error of the finite difference quotients all have a minimum, i.e. that they only have a restricted accuracy. At this level, the non-linear $\mathcal{O}(\Delta x)$ (for forwards and backwards) and also the non-quadratic $\mathcal{O}(\Delta x^2)$ (central) orders of the exponential function start to play a significant role. The solution therefore looses accuarcy for smaller stepsizes.
